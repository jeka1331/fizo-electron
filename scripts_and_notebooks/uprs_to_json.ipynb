```python
allUprsStandards = pd.DataFrame(columns=['uprazhnenieId','categoryId', 'value', 'result'])
```

```python
def cat_name_to_index_mans(name: str):
    if name == '1-я возрастная группа, курсанты 3-го курса и старше':
        return 1
    if name == '1-явозрастная группа, курсанты 3-го курса и старше':
        return 1
    if name == '2- я возрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '2-явозрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '2-я возрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '3-я возрастная группа':
        return 3
    if name == '3-явозрастная группа':
        return 3
    if name == '4-я возрастная группа':
        return 4
    
    if name == '3- я возрастная группа':
        return 3
    if name == '4- я возрастная группа':
        return 4
    if name == '5- я возрастная группа':
        return 5
    if name == '6- я возрастная группа':
        return 6
    if name == '7- я возрастная группа':
        return 7
    if name == '8- я возрастная группа':
        return 8
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву до 6 месяцев службы':
        return 17
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву после 6 месяцев службы':
        return 18
    if name == 'Юноши':
        return 19
    return None

def cat_name_to_index_sfpis(name: str):
    if name == '1-я возрастная группа':
        return 20
    if name == '2- я возрастная группа':
        return 21
    if name == '3- я возрастная группа':
        return 22
    if name == '4- я возрастная группа':
        return 23
    if name == '5- я возрастная группа':
        return 24
    if name == '6- я возрастная группа':
        return 25
    if name == '7- я возрастная группа':
        return 26
    if name == '8- я возрастная группа':
        return 27

    return None
    
```

```python
data = pd.read_excel('../uprs/Упражнение № 3 Подтягивание на перекладине.xls', engine="xlrd",usecols= 'B:U', skiprows=[0], nrows=9).dropna(axis=1) 
data.replace('-', None, inplace=True)  # Заменяем "-" на NaN
data

datar = pd.read_excel('../uprs/Упражнение № 3 Подтягивание на перекладине.xls', engine="xlrd",usecols= 'A', skiprows=[0], nrows=9).dropna(axis=1) 
datar.replace('-', None, inplace=True)  # Заменяем "-" на NaN
datarl = {}
for index, row in datar.iterrows():
    datarl[index] = row.to_list()[0]
    # print(index, row.to_list()[0])

for index, row in data.iterrows():
    # print({index: row.to_dict()})
    for _i , _r in row.to_dict().items():
        category_temp = cat_name_to_index_mans(_i) if cat_name_to_index_mans(_i) is not None else _i
        if (_r) is not None:
            res = (1, category_temp, _r, datarl[index])
            allUprsStandards.loc[len(allUprsStandards)] = res
```
