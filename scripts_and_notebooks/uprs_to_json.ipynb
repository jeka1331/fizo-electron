```python
def cat_name_to_index_mans(name: str):
    if name == '1-я возрастная группа, курсанты 3-го курса и старше':
        return 1
    if name == '1-явозрастная группа, курсанты 3-го курса и старше':
        return 1
    if name == '2- я возрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '2-явозрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '2-я возрастная группа, курсанты и кадеты 2-го курса':
        return 2
    if name == '3-я возрастная группа':
        return 3
    if name == '3-явозрастная группа':
        return 3
    if name == '4-я возрастная группа':
        return 4
    
    if name == '3- я возрастная группа':
        return 3
    if name == '4- я возрастная группа':
        return 4
    if name == '5- я возрастная группа':
        return 5
    if name == '6- я возрастная группа':
        return 6
    if name == '7- я возрастная группа':
        return 7
    if name == '8- я возрастная группа':
        return 8
    if name == '8-я возрастная группа':
        return 8
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву до 6 месяцев службы':
        return 17
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву после 6 месяцев службы':
        return 18
    if name == 'Юноши':
        return 19
    return None


def cat_name_to_index_women(name: str):
    if name == '1- я возрастная группа':
        return 9
    if name == '2- я возрастная группа':
        return 10
    if name == '3- я возрастная группа':
        return 11
    if name == '4- я возрастная группа':
        return 12
    if name == '5- я возрастная группа':
        return 13
    if name == '6 я возрастная группа':
        return 14
    if name == '7- я возрастная группа':
        return 15
    if name == 'Курсанты и кадеты до 6 месяцев службы':
        return 53
    if name == 'Курсанты и кадеты после 6 месяцев службы':
        return 54 
    if name == 'Курсанты и кадеты 2-го курса':
        return 
    if name == 'Курсанты 3-го курса и старше':
        return 
    if name == 'Девушки':
        return 52
    return None
def cat_name_to_index_sfpis(name: str):
    if name == '1-я возрастная группа':
        return 20
    if name == '2- я возрастная группа':
        return 21
    if name == '3- я возрастная группа':
        return 22
    if name == '4- я возрастная группа':
        return 23
    if name == '5- я возрастная группа':
        return 24
    if name == '6- я возрастная группа':
        return 25
    if name == '7- я возрастная группа':
        return 26
    if name == '8- я возрастная группа':
        return 27

    return None
    
def cat_name_to_index_sfpis_15(name: str):
    if name == '1-я возрастная группа до 70 килограмм':
        return 28
    if name == '1-я возрастная группа свыше 70 килограмм':
        return 39
    if name == '2- я возрастная группа до 70 килограмм':
        return 29
    if name == '2- я возрастная группа свыше 70 килограмм':
        return 40
    if name == '3- я возрастная группа до 70 килограмм':
        return 30
    if name == '3- я возрастная группа свыше 70 килограмм':
        return 41
    if name == '4- я возрастная группа до 70 килограмм':
        return 31
    if name == '4- я возрастная группа свыше 70 килограмм':
        return 42
    if name == '5- я возрастная группа до 70 килограмм':
        return 32
    if name == '5- я возрастная группа свыше 70 килограмм':
        return 43
    if name == '6- я возрастная группа до 70 килограмм':
        return 46
    if name == '6- я возрастная группа свыше 70 килограмм':
        return 44
    if name == '7- я возрастная группа до 70 килограмм':
        return 47
    if name == '7- я возрастная группа свыше 70 килограмм':
        return 45

    return None

def cat_name_to_index_mans_15(name: str):
  
    if name == '1-я возрастная группа, курсанты 3-го курса и старше до 70 килограмм':
        return 33
    if name == '1-я возрастная группа, курсанты 3-го курса и старше свыше 70 килограмм':
        return 36
    if name == '2-я возрастная группа, курсанты и кадеты 2-го курса до70 килограмм':
        return 34
    if name == '2-я возрастная группа, курсанты и кадеты 2-го курса свыше 70 килограмм':
        return 37
    if name == '3-я возрастная группа группы до 70 килограмм':
        return 35
    if name == '3-я возрастная группа группы свыше 70 килограмм':
        return 38
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву до 6 месяцев службы до 70 кило грамм':
        return 48
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву до 6 месяцев службы свыше 70 килограмм':
        return 50
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву после 6 месяцев службы до 70 килограмм':
        return 49
    if name == 'Курсанты, кадеты и военнослужащие, проходящие воинскую службу по призыву после 6 месяцев службы свыше 70 килограмм':
        return 51


    return None
```
